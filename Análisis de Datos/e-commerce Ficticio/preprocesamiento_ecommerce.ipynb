#Preprocesamiento del archivo 'customers.csv'

```
df_customers = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/fuentes/customers.csv')
```

```
df_customers
```

```
df_customers_copia = df_customers
```

```
df_customers_copia
```

```
df_customers_copia.info()
```

La columna "Gender" tiene datos faltantes.

```
df_customers_copia.duplicated(subset=['Name','Surname','Gender','Birthdate']).sum()
```

Se detectan 158 registros como duplicados.

```
customers_duplicated = df_customers_copia.duplicated(['Name','Surname','Gender','Birthdate'])[df_customers_copia.duplicated(['Name','Surname','Gender','Birthdate']) == True].index
```

```
customers_duplicated
```

```
df_customers_copia = df_customers_copia.drop(customers_duplicated,axis=0)
```

```
df_customers_copia
```

```
df_customers_copia.reset_index(drop=True, inplace=True)
```

```
df_customers_copia
```

```
df_customers_copia.duplicated(subset=['Name','Surname','Gender','Birthdate']).sum()
```

Los registros duplicados ya fueron eliminados. Ahora queda transformar la columna de "Gender".

No se eliminaran los registros con el valor "NaN" en "Gender" porque es importante tenerlos en cuenta en el anÃ¡lisis. Se reemplazarÃ¡ el valor faltante por un string.

```
df_customers_copia['Gender'].replace(np.NaN,'Sin especificar',inplace=True)
```

```
df_customers_copia
```

```
df_customers_copia.info()
```

Ya no quedan registros faltantes ni duplicados.

Se exporta el DataFrame.

```
df_customers_copia.to_csv('/content/drive/MyDrive/Colab Notebooks/fuentes/customers_clean.csv', index=False)
```
